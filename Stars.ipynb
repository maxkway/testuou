## Distant stars..
Here are solved some task about measurement angel's of very distant object by using interference.
![Scheme](https://sun9-4.userapi.com/LjudTBUtzcLdRbcTN15YgG3iem0_mTID_1PMIg/XHNCdA99JgE.jpg)
It is easy to see, that result field in the summator is:
$$
E\sim e^{ikr_1} e^{ikr_2} \implies I \sim 2 I_0 \left[1 + \cos\left(\frac{2\pi}{\lambda}\delta r\right)\right]
$$
And we can roughly estimate :
$$
\delta r = B \sin \theta
$$
We got dimensionless parameter:
$$
\xi = \frac{2\pi}{\lambda} B \sim 5 \cdot10^{3} \gg 1
$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
l = 0.50 # m
B = 480 # m

xi = 2*np.pi/l*B
print(1/xi)
```

```python
xi = 100
```

```python
%matplotlib notebook
t = np.linspace(0, np.pi/2, 5000)
i = 1 + np.cos(xi*np.sin(t))
plt.figure()
plt.plot(t, i)
plt.title(r'Intensity from $\theta$, $\xi = 100$')
plt.grid()
plt.xticks(np.linspace(0,np.pi/2,4), 
          (r'0' , r'$\frac{\pi}{6}$',r'$\frac{\pi}{3}$',r'$\frac{\pi}{2}$'))
```

As we can see, it is good idea to measure this ways on some angle..
Lets plot a derivation of intensivety:
$$
\delta I \sim sin(\xi \sin \theta )\xi \cos \theta\, \delta \theta
$$

```python
%matplotlib notebook
di = np.sin(xi*np.sin(t))*np.cos(t)
plt.figure()
plt.plot(t, di)
plt.title(r'Derivation of intensity from $\theta$, $\xi = 100$')
plt.grid()
plt.xticks(np.linspace(0,np.pi/2,4), 
          (r'0' , r'$\frac{\pi}{6}$',r'$\frac{\pi}{3}$',r'$\frac{\pi}{2}$'))
```

**Thus $\theta$ is measuremed by $1/\xi$**

### Non single waves.
Anfortunately, in the real life we have to use signales with wight $\delta \lambda$ of cpectrum.
In this case We have to intagrate gotten result for single $\lambda$ with known weight $\rho (\lambda)$:
$$
I_{\sum}\sim \int \rho(\lambda) I_{1}(x,\lambda) \text{ d} \lambda
$$

But it works good only with $$ \frac{\delta \lambda}{\lambda} \ll 1$$
To rought estimate influence on the measurements we can use distibution on the frecuency, and this weight:
$$
\rho(\omega) =\cases{1, & $|\omega-\omega_0| < \frac{\delta \omega}{2}$ \\
                     0, & otherwise}
$$

We get fast-oscilating previous function with slow moduling function near constant:
$$
I_{\sum}\sim 1 + \left[\frac{\sin (\theta \frac{B \delta \omega  }{2c})}{\theta \frac{B \delta \omega  }{2c}} \right]\cdot \cos \left(\theta \frac{B}{c}\omega\right)
$$
Is is
$$
I =1 + \frac{\sin x}{x}\cdot \cos\left(x\, \frac{\lambda_0}{\delta \lambda}\right)
$$

```python
x = np.linspace(0.01,20,5000)
iw = 1 + np.sin(x)/ x * np.cos(40*x)
plt.figure()
plt.plot(x, iw,label='Intyncity')
isin = 1 + np.sin(x)/ x
plt.plot(x, isin,label=r'Modulation $1 + \frac{\sin(x)}{x}$')
plt.title(r'Intencity with $\delta \lambda$')
plt.hlines(1.1, 0, 20, colors='r', linestyles='--', label='Limit of sensitive')
plt.hlines(0.9, 0, 20, colors='r', linestyles='--')
plt.annotate(r'$\delta I(\theta > \theta^{*})< 0.1 $', xy=(11, 1.1),
            xycoords='data', xytext=(40, 10), textcoords='offset points',
            fontsize=16,
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
plt.legend()
plt.grid()
```
