# Monte-Carlo Error

$F = G\frac{M_1 M_2}{r^2}$ , $G = 6,67384 \times 10^{-11}$

## First data
$M_1 = (40 \pm 0.05)\times 10^4$ kg

$M_2 = (30 \pm 0.1)\times 10^4$ kg

$r = (3.2 \pm 0.01)$ m

```python
import matplotlib.pyplot as plt
G = 6.67384
m_1 = np.random.normal(40, 0.05, 1000000)
m_2 = np.random.normal(30, 0.1, 1000000)
r = np.random.normal(3.2, 0.01, 1000000)

f = G * m_1 * m_2 / r**2

M_1 = 40; M_2 = 30; R = 3.2
F = G * M_1 * M_2 / R**2

sqrsigma = F**2 * ((0.05/M_1)**2 + (0.1/M_2)**2 + 4 * (0.01/R)**2)

x = np.linspace(760,800,1000)
p = 1 / np.sqrt(2 * np.pi * sqrsigma) * np.exp(-(F - x)**2 / 2 / sqrsigma)

n, bins, patches = plt.hist(f, 50, density=True,histtype='stepfilled')
plt.xlabel('Forse, mH')
plt.ylabel('pcd')
plt.title('Histogram of forse 1')
plt.text(763, .05, r'$\mu=782,\ \sigma=31$', fontsize=14)
plt.axis([760, 803, 0, 0.075])
plt.plot(x,p,markersize=10, linewidth=3,label='pdf')
plt.grid(True)
##plt.show()
```

```python
print(f.mean(), f.std()**2)
```
